I started with the question, “Are first babies more likely to be late?” 

To address it, I computed the difference in means between groups of babies, but I ignored the possibility that there might be a difference between first babies and others for the same woman.

To address this version of the question, select respondents who have at least two babies and compute pairwise differences. Does this formulation of the question yield a different result?

Hint: use nsfg.MakePregMap.

```python
import nsfg
%matplotlib inline
```

```python
preg = nsfg.ReadFemPreg()
live = preg[preg.outcome == 1]
```

```python
live = live[live.prglngth >= 37]
preg_map = nsfg.MakePregMap(live)
type(preg_map)
```

```python
def Diffs(t):
    """List of differences between the first elements and others.
    t: list of numbers    
    returns: list of numbers
    """
    first = t[0]
    rest = t[1:]
    diffs = [first - x for x in rest]
    return diffs
```

```python
diffs = []
for caseid, indices in preg_map.items():
    lengths = live.loc[indices].prglngth.values
    # check differences between 1 and other babies
    if len(lengths) >= 2:
        diffs.extend(Diffs(lengths))
```

```python
import thinkstats2
mean = thinkstats2.Mean(diffs)
print 'Mean difference between pairs: {}'.format(mean)
```

```python
pmf = thinkstats2.Pmf(diffs)
```

```python
import thinkplot
thinkplot.Hist(pmf, align='center')
thinkplot.Show(xlabel='Difference in weeks', ylabel='PMF')
```

Looking at the plot, we see there is a slightly higher bars on the positive side, meaning first babies do tend to arrive eralier when looking at pairs of babies of the same mother.
